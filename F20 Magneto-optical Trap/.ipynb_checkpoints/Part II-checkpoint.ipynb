```python
def grabData(file):
    #file is string with form : '<absolute detuning value>_B<current value>/F00<# of csv>'
    dg = np.genfromtxt(file, delimiter=',')
    g1,g2,g3,g4,g5,g6 = np.hsplit(dg,6)
    #return g4 = time array, g5 = voltage value
    return g4,g5
def con(x,b):
    return 0*x+b
```
